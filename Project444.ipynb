```python
# แสดงผล top 10 ประโยคที่คล้ายกันที่สุด
def find_similar_sentences(sentence_embeddings, similarities, top_k=10):
  results = []
  for i in range(len(sentence_embeddings)):
      scores, indices = torch.topk(similarities[i], k=top_k + 1)  # รวมประโยคตัวเองด้วย
      #print(f"\nTop {top_k} similar sentences for Sentence {i+ 1} {df_combined['Name'][i]}:")
      for score, idx in zip(scores[1:], indices[1:]):  # ข้ามตัวเอง
          #print(f"Sentence {idx.item() + 1} {df_combined['Name'][idx.item()]} (Score: {score.item():.4f})")
          results.append({
                "Source_Place": df['Name'][i],         # ประโยคต้นทาง
                "Similar_Place": df['Name'][idx.item()],  # ประโยคที่คล้ายกัน
                "Similarity_URL_Place" : df['Image_URL'][idx.item()],  #รูปประโยคที่คล้ายกัน
                "Similarity Score": score.item()                      # คะแนนความคล้ายกัน
            })

  # แสดงผลลัพธ์ในรูปแบบ DataFrame
  df_results = pd.DataFrame(results)
  return df_results
```

```python
find_similar_sentences(sentence_embeddings, similarities, top_k=10)
```

```python
# หาประโยคที่ต่างกันมากที่สุด (ค่าความคล้ายคลึงต่ำสุด)
def find_dissimilar_sentences(sentence_embeddings, similarities, top_k=10):
  results = []
  for i in range(len(sentence_embeddings)):
    top_dissimilar_scores, top_dissimilar_indices = torch.topk(-similarities[i], k=top_k)  # ใช้ค่าติดลบเพื่อหาค่าต่ำสุด
    #print(f"\nTop {top_k} dissimilar sentences for Sentence {i + 1} {df_combined['Name'][i]}:")
    for score, idx in zip(-top_dissimilar_scores, top_dissimilar_indices):  # กลับด้านค่าคะแนนเพื่อให้เป็นบวก
        #print(f"Sentence {idx.item() + 1} {df_combined['Name'][idx.item()]} (Score: {score.item():.4f})")
          results.append({
                "Source_Place": df['Name'][i],         # ประโยคต้นทาง
                "Dissimilar_Place": df['Name'][idx.item()],  # ประโยคที่ต่างกัน
                "Dissimilarity_URL_Place" : df['Image_URL'][idx.item()],  #รูปประโยคที่ต่างกัน
                "Dissimilarity Score": score.item()                      # คะแนนความต่างกัน
            })

  # แสดงผลลัพธ์ในรูปแบบ DataFrame
  df_results = pd.DataFrame(results)
  return df_results
```

```python
# สร้างกราฟ
G = nx.Graph()

# เพิ่มโหนด (ข้อความ)
for i, text in enumerate(texts):
    G.add_node(i, text=text)

# เพิ่ม edges จากความคล้ายกัน (เช่น เอาเฉพาะ similarity ที่มากกว่า 0.5)
threshold = 0.5
for i in range(len(texts)):
    for j in range(i + 1, len(texts)):
        if similarities[i, j] > threshold:
            G.add_edge(i, j)

# แสดงกราฟ
print(G.edges)
```

```python
def find_top_similarities(similarities, top_k=10):
    top_similarities = []
    for i in range(len(similarities)):
        scores, indices = torch.topk(similarities[i], k=top_k + 1)
        top_similarities.append((scores[1:], indices[1:]))  # ข้ามตัวเอง
    return top_similarities
```

```python
# Top 10 similarities สำหรับทั้ง Semantic และ Graph Embeddings
semantic_top_similarities = find_top_similarities(similarities, top_k=10)
graph_top_similarities = find_top_similarities(node_similarities, top_k=10)

# แสดงผลเปรียบเทียบ
all_mean_semantic_similarities =[]
all_mean_graph_similarities = []
for i in range(len(semantic_top_similarities)):
    # print(f"\nSentence {i+1}")
    # print("Semantic Top Similarities:", torch.mean(semantic_top_similarities[i][0]))
    # print("Graph Top Similarities:", torch.mean(graph_top_similarities[i][0]))
    all_mean_semantic_similarities.append(torch.mean(semantic_top_similarities[i][0]))
    all_mean_graph_similarities.append(torch.mean(graph_top_similarities[i][0]))

print("All_mean_semantic",torch.mean(torch.tensor(all_mean_semantic_similarities)))
print("All_mean_graph",torch.mean(torch.tensor(all_mean_graph_similarities)))
```

```python
def semantic_plot_tags(semantic_embedding_reduced, tags):
    plt.figure(figsize=(20, 10))

    for tag in tags_english_map.keys():
        tag_points = semantic_embedding_reduced[tags == tag]

        if len(tag_points) > 1:  # ต้องมีจุดมากกว่าหนึ่งจุด
            # Plot จุด
            plt.scatter(tag_points[:, 0], tag_points[:, 1],
                        s=100, color=tag_colors[tag],
                        alpha=0.7, label=tags_english_map[tag])

    plt.title('Semantic Embedding with Tags')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.legend(title='Tags Legend')
    plt.grid(True)
    plt.show()
```

```python
def semantic_plot_zoning(semantic_embedding_reduced, semantic_labels, zoning):
    unique_clusters = np.unique(semantic_labels)
    cluster_colors = plt.cm.get_cmap('viridis', len(unique_clusters))

    plt.figure(figsize=(20, 10))

    for cluster in unique_clusters:
        cluster_points = semantic_embedding_reduced[semantic_labels == cluster]
        cluster_zones = zoning[semantic_labels == cluster]

        for zone in zoning_english_map.keys():
            zone_points = cluster_points[cluster_zones == zone]

            # Plot จุด
            plt.scatter(
                zone_points[:, 0], zone_points[:, 1],
                s=100,
                marker=zoning_marker_map[zone],  # ใช้ marker ต่างกันตาม zoning
                color=cluster_colors(cluster),
                alpha=0.7,
                label=f'{zoning_english_map[zone]} (Cluster {cluster})'
            )

    plt.title('Semantic Embedding with Zoning')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.legend(title='Zoning Legend')
    plt.grid(True)
    plt.show()
```
